# Importing Libraries

```python
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
```

```python
import pandas as pd
```

```python
from google.colab import userdata
Google_sec_key=userdata.get("GOOGLEAPI")
```

```python
from langchain_google_genai import GoogleGenerativeAI, HarmBlockThreshold, HarmCategory
llm = GoogleGenerativeAI(
    model="gemini-pro",
    google_api_key=Google_sec_key,
    temperature=0.5,
    safety_settings={
        HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: HarmBlockThreshold.BLOCK_NONE,
    },
)
```

```python
url = "https://raw.githubusercontent.com/adamerose/datasets/master/titanic.csv"
df=pd.read_csv(url)
print(df.shape)
df.head()
```

```python
from langchain_huggingface import HuggingFaceEndpoint
```

```python
from google.colab import userdata
sec_key=userdata.get("sec_key")
```

```python
llm2=HuggingFaceEndpoint(
    model="mistralai/Mistral-7B-v0.1",
    task="text-generation",
    huggingfacehub_api_token=sec_key,
    timeout=300  # Increase timeou
)
```

# Single DatFrame Example
#Agent can interact with Single DataFRame

```python
agent=create_pandas_dataframe_agent(llm2,df,verbose=True,allow_dangerous_code=True)
```

```python
agent.run("How many rows are there?")
```

```python
agent.run("How many people have more than 23 age")
```
